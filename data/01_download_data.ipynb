```python
import requests

# Assuming you have an API token if needed for accessing your files
api_token = "a0d3e5db-27c2-4177-ab28-70a45c40b3db"

# Depending on your permissions, you might need to include the API token in the header
headers = {
    'X-Dataverse-key': api_token
}

# Replace 'your_dataverse_identifier' with your specific dataverse identifier
base_url = "https://dataverse.harvard.edu/dataset.xhtml?persistentId=doi:10.7910/DVN/MUEGDT&version=25.0"

# Get the list of datasets in your dataverse
datasets_response = requests.get(base_url, headers=headers)
```

```python

# Check if the request was successful
if datasets_response.status_code == 200:
    datasets = datasets_response.json()['data']
    for dataset in datasets:

        # Get the dataset ID or persistent identifier here
        # Depending on the Dataverse set-up, you might need to use a different key
        dataset_id = dataset['id'] # or dataset['persistentId'] if using a persistent identifier

        # Construct the URL to access files within a dataset
        files_url = f"https://dataverse.harvard.edu/api/datasets/:persistentId/versions/:latest/files?persistentId={dataset_id}"

        # Get the list of files in the dataset
        files_response = requests.get(files_url, headers=headers)

        if files_response.status_code == 200:
            files = files_response.json()['data']

            for file in files:
                # Extract the download URL for each file
                download_url = file['dataFile']['downloadUrl']

                # Download the file
                file_response = requests.get(download_url, headers=headers)
                if file_response.status_code == 200:
                    # Write the file to the local system
                    filename = file['dataFile']['filename']
                    with open(filename, 'wb') as f:
                        f.write(file_response.content)
                    print(f"Downloaded {filename}")
                else:
                    print(f"Failed to download file {file['dataFile']['filename']}")
else:
    print("Failed to retrieve datasets")

# Note: Error handling should be improved for production code.
```
