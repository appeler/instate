### Cleaning names

```python
# Regular expression pattern to match valid names: alphabetic characters, spaces, hyphens, and apostrophes
pattern = r"^[a-zA-Z\s'-]+$"

# Filter the DataFrame to include only valid names in 'name1'
df = df[df['elector_name'].str.contains(pattern)]
df = df[df['father_or_husband_name'].str.contains(pattern, na=False)]
```

Given are full names, so split and take last names

```python
# Function to normalize names and return a set of possible last names
def get_possible_last_names(name):
    parts = name.split(' ')
    return set(parts)

# Apply the function to each name column
df['last_names1'] = df['elector_name'].apply(get_possible_last_names)
df['last_names2'] = df['father_or_husband_name'].apply(get_possible_last_names)
```

By find the match between elector name and father or husband name we are getting last name match

```python
# Find matching rows by checking for intersections in the sets of last names
df['last_name_match'] = df.apply(lambda row: len(row['last_names1'].intersection(row['last_names2'])) > 0, axis=1)
```

```python
df.shape
```

```python
df = df[df['last_name_match']]
```

```python
# Helper function to extract last names and filter out invalid ones
def extract_valid_last_name(name1, name2):
    # Split the names by space
    parts1 = name1.split()
    parts2 = name2.split()

    # Extract the last names
    last_name1 = parts1[-1] if len(parts1) > 1 else ''
    last_name2 = parts2[-1] if len(parts2) > 1 else ''

    # Check if the last names are valid (not a single or double letter) and match
    if last_name1 == last_name2 and len(last_name1) > 2:
        return last_name1
    else:
        return None
```

```python
# Apply the helper function to each row
df['last_name'] = df.apply(lambda row: extract_valid_last_name(row['elector_name'], row['father_or_husband_name']), axis=1)

# Filter out the rows where last_name is None
df = df.dropna(subset=['last_name'])
```

```python
def process_lastnames(file, state):
    df = pd.read_csv(file)
    # Regular expression pattern to match valid names: alphabetic characters, spaces, hyphens, and apostrophes
    pattern = r"^[a-zA-Z\s'-]+$"

    # drop na values in columns elector_name and father_or_husband_name
    df = df.dropna(subset=['elector_name', 'father_or_husband_name'])

    # Filter the DataFrame to include only valid names in 'name1'
    df = df[df['elector_name'].str.contains(pattern)]
    df = df[df['father_or_husband_name'].str.contains(pattern, na=False)]

    # Function to normalize names and return a set of possible last names
    def get_possible_last_names(name):
        parts = name.split(' ')
        return set(parts)

    # Apply the function to each name column
    df['last_names1'] = df['elector_name'].apply(get_possible_last_names)
    df['last_names2'] = df['father_or_husband_name'].apply(get_possible_last_names)

    # Find matching rows by checking for intersections in the sets of last names
    df['last_name_match'] = df.apply(lambda row: len(row['last_names1'].intersection(row['last_names2'])) > 0, axis=1)

    # Helper function to extract last names and filter out invalid ones
    def extract_valid_last_name(name1, name2):
        # Split the names by space
        parts1 = name1.split()
        parts2 = name2.split()

        # Extract the last names
        last_name1 = parts1[-1] if len(parts1) > 1 else ''
        last_name2 = parts2[-1] if len(parts2) > 1 else ''

        # Check if the last names are valid (not a single or double letter) and match
        if last_name1 == last_name2 and len(last_name1) > 2:
            return last_name1
        else:
            return None

    # Apply the helper function to each row
    df['last_name'] = df.apply(lambda row: extract_valid_last_name(row['elector_name'], row['father_or_husband_name']), axis=1)

    # Filter out the rows where last_name is None
    df = df.dropna(subset=['last_name'])

    # Convert the last_name column to lowercase
    df['last_name'] = df['last_name'].str.lower()

    # Get the value counts of the last names
    last_name_counts = df['last_name'].value_counts()

    # Filter the counts to keep only those greater than 50
    last_name_counts_filtered = last_name_counts[last_name_counts > 10]

    # Convert the Series to a DataFrame
    last_name_counts_df = last_name_counts_filtered.reset_index()
    last_name_counts_df.columns = ['last_name', 'count']

    # Add the state column
    last_name_counts_df['state'] = state

    # Save the DataFrame to a CSV file
    last_name_counts_df.to_csv(f'data/clean/{state}_ln.csv', index=False)
```

```python
def merge_agg(files, state):
    dataframes = [pd.read_csv(file) for file in files]
    combined_df = pd.concat(dataframes, ignore_index=True)

    aggregated_data = combined_df.groupby('last_name')['count'].sum().reset_index()
    aggregated_data.columns = ['last_name', 'count']
    aggregated_data['state'] = state

    aggregated_data.to_csv(f'data/non_english/{state}_ln.csv', index=False)
```
